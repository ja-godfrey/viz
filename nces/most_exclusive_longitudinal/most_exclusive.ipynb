```python
names = pd.DataFrame()
for proj in csvs:
    frame = pd.read_csv(proj, usecols=['INSTNM', 'UGDS']).dropna()
    frame['year'] = int(proj[-14:-10])
    names = names.append(frame, ignore_index=True)
```

```python
names
```

```python
x = pd.pivot_table(names, values='UGDS', index=['INSTNM'], columns='year', aggfunc=np.sum)
x
```

```python
z = x.T
```

```python
z
```

```python
z.fillna(0, inplace=True)
z.sort_values(list(z.columns), inplace=True)
z = z.sort_index()
z.head(5)
```

```python
top_prem_clubs = set()

for index, row in z.iterrows():
    top_prem_clubs |= set(row[row > 0].sort_values(ascending=False).head(20).index)

z = z[top_prem_clubs]
```

```python
z
```

```python
#further cleaning
z['University of Pheonix'] = z['University of Phoenix-Arizona'] + z['University of Phoenix-Online Campus']
z.drop(["University of Phoenix-Arizona", "University of Phoenix-Online Campus"], 1, inplace=True)

#delete 0 dudes
z = z.replace(0,np.nan).dropna(axis=1,how="any")
```

```python
z = z.T
```

```python
z[2021]=z[2020]
```
