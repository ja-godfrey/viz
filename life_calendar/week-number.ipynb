# What week number is a certain date?

To backfill the life calendar, it is useful to have a quick way to calculate which week # any given date occured in. For example, what week # is June 18<sup>th</sup> 1992?

```python
import datetime
def week_no(y, m, d):
    return datetime.date(y, m, d).isocalendar()[1]
```

```python
week_no(2022, 6, 26)
```
