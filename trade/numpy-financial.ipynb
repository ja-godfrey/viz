## sensitivity analysis

```python
def future_value(rate, periods, initial_payment, principal_value):
    fv = npf.fv(rate, periods, -initial_payment, -principal_value)
    return fv
```

```python
future_value(.07, 5, 10000, 1000)
```

```python
def future_value_variedRates(rates, periods, initial_payment, principal_value):
    fv = npf.fv(rates, periods, -initial_payment, -principal_value)
    z = pd.DataFrame(fv, rates, columns=['Ending Value'])
    return z
```

```python
rates = [rate / 100 for rate in range (3,10)]
future_value_variedRates(rates, 5, 10000, 1000)
```

```python
rate = .07
periods = [i for i in range (1,100)]
pmt = 10000
pv = 30000
fv = npf.fv(rate, periods, -pmt, -pv)
pd.DataFrame(fv, periods, columns=['Ending Value'])
```
