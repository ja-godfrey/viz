```python
import pandas as pd
import numpy as np
from itertools import chain 
```

```python
df = pd.read_csv('spotifyPlaylists.csv', engine='python')
df = df.drop(['Date Added'], axis=1)
df_2020 = df[df.Year == 2020]
```

```python
df.head()
```

```python
df.dtypes
```

```python
# most popular artists
popArtist = df.pivot_table(index=['Artist'], aggfunc='size')
popArtist.sort_values(ascending=False, inplace=True)
print(popArtist[:10])
```

```python
# most popular artists 2020
popArtist = df_2020.pivot_table(index=['Artist'], aggfunc='size')
popArtist.sort_values(ascending=False, inplace=True)
print(popArtist[:10])
```

```python
#most popular albums 2020
popArtist = df_2020.pivot_table(index=['Album'], aggfunc='size')
popArtist.sort_values(ascending=False, inplace=True)
print(popArtist[:10])
```
